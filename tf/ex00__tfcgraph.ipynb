```
import tensorflow as tf
```

```
a = tf.placeholder(tf.float64, name='a')
na = tf.placeholder(tf.float64, name='na')
b = tf.placeholder(tf.float64, name='b')
nb = tf.placeholder(tf.float64, name='nb')
c = tf.placeholder(tf.float64, name='c')
#aa = tf.multiply(na, a, name='aa')
aa = na * a
#bb = tf.multiply(nb, b, name='bb')
bb = nb * b
#aabb = tf.add(aa, bb, name='aabb')
aabb = aa + bb
#caabb = tf.multiply(c, aabb, name='caabb')
caabb = c * aabb
```

```
ga, gna, gb, gnb, gaa, gbb, gaabb, gc, gcaabb = tf.gradients(caabb, [a, na, b, nb, aa, bb, aabb, c, caabb])
```

```
feed_dict = {a: 100, na: 2, b: 150, nb: 3, c: 1.1}
```

```
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  v_caabb = sess.run(caabb, feed_dict)
  print(v_caabb)
  grads = sess.run([ga, gna, gb, gnb, gaa, gbb, gaabb, gc, gcaabb], feed_dict)
  print(grads)
```

Soft thresholding

```
def soft_thresholding(r_, lam_):
    "implement a soft threshold function y=sign(r)*max(0,abs(r)-lam)"
    lam_ = tf.maximum(lam_, 0)
    return tf.sign(r_) * tf.maximum(tf.abs(r_) - lam_, 0)
```

```
z = tf.placeholder(tf.float64, name='z')
th = tf.placeholder(tf.float64, name='th')
soft_z = soft_thresholding(z, th)
g = tf.gradients(soft_z, [z, th])
```

```
import numpy as np
feed_dict = {z: np.linspace(-3., 3., 21, dtype=np.float64), th: 0.5}
sess = tf.Session()
sess.run(soft_z, feed_dict)
```

```
#feed_dict = {z: np.linspace(-3., 3., 21, dtype=np.float32), th: 0.5}
feed_dict = {z: np.linspace(-3., 3., 21, dtype=np.float32), th: np.ones(21)*0.5}
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  v_soft_z = sess.run(soft_z, feed_dict)
  print(v_soft_z)
  grads = sess.run(g, feed_dict)
  print(grads)
```

問題　連立方程式

x+y+z=0
2x+y+3z+-5
-x+3x-z=8

```
mtr1 = tf.placeholder(tf.float32, shape=(3,3))
mtr2 = tf.placeholder(tf.float32, shape=(3,1))
```

```
inverse = tf.matrix_inverse(mtr1)
mat = tf.matmul(inverse,mtr2)
```

```
with tf.Session() as sess:
  result = sess.run(mat, feed_dict={mtr1:[[1.,1.,1.],[2.,1.,3.],[-1.,3.,-1.]], mtr2:[[0.],[-5.],[8.]]})
  print(result)
```
